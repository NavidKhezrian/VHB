```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam

# Define the input data (flattened 2x2 grayscale image)
x_train = np.array([[0.2, 0.7, 0.3, 0.9]])

# Define the autoencoder architecture
input_dim = 4  # Size of input layer (2x2 image flattened)
encoding_dim = 2  # Size of encoded layer (latent space)

# Define the encoder model
input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)

# Define the decoder model
decoded = Dense(input_dim, activation='sigmoid')(encoded)

# Create the autoencoder model
autoencoder = Model(inputs=input_layer, outputs=decoded)

# Compile the model
autoencoder.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')

# Train the autoencoder
autoencoder.fit(x_train, x_train, epochs=500, verbose=0)

# Define the encoder model to extract encoded features
encoder = Model(inputs=input_layer, outputs=encoded)

# Encode the input data
encoded_data = encoder.predict(x_train)

# Decode the encoded data
decoded_data = autoencoder.predict(x_train)

# Extract weights and biases
encoder_weights = encoder.layers[1].get_weights()
decoder_weights = autoencoder.layers[2].get_weights()

print("Original Input Data:")
print(x_train)

print("Encoded Data:")
print(encoded_data)

print("Decoded Data (Reconstructed Input):")
print(decoded_data)

print("\nEncoder Weights:")
print("Weights:")
print(encoder_weights[0])
print("Biases:")
print(encoder_weights[1])

print("\nDecoder Weights:")
print("Weights:")
print(decoder_weights[0])
print("Biases:")
print(decoder_weights[1])
```

```python
import numpy as np

# Define input data
x = np.array([0.2, 0.7, 0.3, 0.9])

# Define encoder weights and biases
W_e = encoder_weights[0]

b_e = encoder_weights[1]

# Define decoder weights and biases
W_d = decoder_weights[0]

b_d = decoder_weights[1]

# Define activation functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Encoding
def encode(x, W_e, b_e):
    linear_transformation = np.dot(x, W_e) + b_e
    return relu(linear_transformation)

# Decoding
def decode(encoded_x, W_d, b_d):
    linear_transformation = np.dot(encoded_x, W_d) + b_d
    return sigmoid(linear_transformation)

# Perform encoding
encoded_data = encode(x, W_e, b_e)
print("Encoded Data:")
print(encoded_data)

# Perform decoding
decoded_data = decode(encoded_data, W_d, b_d)
print("Decoded Data (Reconstructed Input):")
print(decoded_data)
```
